```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
kmeans = KMeans(n_clusters=240, init='k-means++')

# Compute the clusters based on longitude and latitude features
X_sample = df[['longitude','latitude']].sample(frac=0.1)
kmeans.fit(X_sample)
y = kmeans.labels_
```

```python
df['cluster'] = kmeans.predict(df[['longitude','latitude']])
df[['name','latitude','longitude','votes','cluster', 'rate', 'review']].sample(10)
```

```python
import numpy as np
def recommend_venues(df, longitude, latitude):
    predicted_cluster = kmeans.predict(np.array([longitude,latitude]).reshape(1,-1))[0]
    # Fetch the venue name of the top most record in the topvenues dataframe for the predicted cluster
    venue_name = df[df['cluster']==predicted_cluster].iloc[0]['name']
    msg = 'As due to covid try visiting the ' + venue_name + '?'
    return msg
```

```python
import string
string.punctuation
```

```python
text_nopunct=''
text_nopunct= "".join([char for char in text_corpus if char not in string.punctuation])
```

```python
tokenizer = nltk.tokenize.RegexpTokenizer('\w+')
```

```python
text_tokens = tokenizer.tokenize(text_nopunct)
len(text_tokens)
```

```python
text_words = []
for word in text_tokens:
    text_words.append(word.lower())
```

```python
nltk.download('stopwords')
```

```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
final_words=[]

for word in text_words:
    if word not in stopwords:
        final_words.append(word)
```

```python
nltk.download('wordnet')
```

```python
wn = WordNetLemmatizer()
```

```python
lem_final_words=[]

for word in final_words:
    word=wn.lemmatize(word)
    lem_final_words.append(word)
```

```python
freq_dist_text = nltk.FreqDist(lem_final_words)
```

```python
plt.subplots(figsize=(20,12))
freq_dist_text.plot(30)
```
